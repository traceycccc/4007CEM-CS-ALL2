```python
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical
import matplotlib.pyplot as plt  # Built-in Matplotlib
import seaborn as sns  # For graphical
from calendar import month_name as mn
import datetime
import sqlite3
connect = sqlite3.connect('Iccountant')
cursor = connect.cursor()
today = datetime.datetime.now()
```

```python
category_in_total_amount = pd.read_sql_query("SELECT ty.type_name AS Type, c.cat_name AS Category, sum(t.amount) \
                                             AS Amount FROM transactions t, category c, type ty WHERE \
                                             c.cat_id = t.cat_id AND t.user_id = c.user_id AND t.type_id = ty.type_id \
                                             AND t.type_id = 1 AND t.user_id = 1 GROUP BY t.cat_id", connect)
dataframe = pd.DataFrame(category_in_total_amount)
print(dataframe)
# type_in_total = dataframe['Type'].values.tolist()
category_in_total = dataframe['Category'].values.tolist()
print(category_in_total)
```

```python
# ================== Bar chart for yearly ==========================
# income yearly
yearlyBarIncome = pd.read_sql_query("SELECT strftime('%m', t.date) AS Month, ty.type_name AS Type, sum(t.amount) AS Amount "
                                "FROM transactions t, user u, type ty WHERE u.user_id = t.user_id AND t.type_id = ty.type_id "
                                "AND t.user_id = 1 AND t.type_id = 1 AND strftime('%Y', t.date) = '2021' GROUP BY "
                                "strftime('%m', t.date)", connect)
dfBarIn = pd.DataFrame(yearlyBarIncome)

monthIn_total = dfBarIn['Month'].values.tolist()
TypeIn_total = dfBarIn['Type'].values.tolist()
AmountIn_total = dfBarIn['Amount'].values.tolist()

# expense yearly
yearlyBarExpense = pd.read_sql_query("SELECT strftime('%m', t.date) AS Month, ty.type_name AS Type, sum(t.amount) AS Amount "
                                      "FROM transactions t, user u, type ty WHERE u.user_id = t.user_id AND t.type_id = ty.type_id "  
                                      "AND t.user_id = 1 AND t.type_id = 2 AND strftime('%Y', t.date) = '2021' GROUP BY "
                                       "strftime('%m', t.date)", connect)
dfBarEX = pd.DataFrame(yearlyBarExpense)

monthEx_total = dfBarEX['Month'].values.tolist()
TypeEx_total = dfBarEX['Type'].values.tolist()
AmountEx_total = dfBarEX['Amount'].values.tolist()

type_total_combine = TypeIn_total + TypeEx_total
month_total_combine = monthIn_total + monthEx_total
amount_total_combine = AmountIn_total + AmountEx_total
df1 = pd.DataFrame(list(zip(month_total_combine, type_total_combine, amount_total_combine)), columns =['Month', 'Type', 'Amount'])

fig, ax = plt.subplots(figsize=(7, 5), dpi=100)
plot = sns.barplot(x='Month', y='Amount', data=df1, hue='Type', palette="Pastel1", ci=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plot.set_ylabel("Amount (RM)", fontsize=15)
plot.set_xlabel("Month", fontsize=15)
plot.set_title("Income and Expense in a Year", fontsize=20)
plot.legend(bbox_to_anchor=(1,1), loc="upper right", borderaxespad=0)
plt.show()
```

```python
# ============================= Bar chart for montly ====================================
# income for monthly
monthlyBarIncome = pd.read_sql_query("SELECT strftime('%d', t.date) AS Day, ty.type_name AS Type, sum(t.amount) AS Amount "
                                "FROM transactions t, user u, type ty WHERE u.user_id = t.user_id AND t.type_id = ty.type_id "
                                "AND t.user_id = 1 AND t.type_id = 1 AND strftime('%Y', t.date) = '2022' AND strftime('%m', t.date) = '10'"
                                "GROUP BY strftime('%d', t.date)", connect)
dfBarInMonth = pd.DataFrame(monthlyBarIncome)

dayIn_total = dfBarInMonth['Day'].values.tolist()
TypeIn_total = dfBarInMonth['Type'].values.tolist()
AmountIn_total = dfBarInMonth['Amount'].values.tolist()

# expense for monthly
monthlyBarExpense = pd.read_sql_query("SELECT strftime('%d', t.date) AS Day, ty.type_name AS Type, sum(t.amount) AS Amount "
                                      "FROM transactions t, user u, type ty WHERE u.user_id = t.user_id AND t.type_id = ty.type_id "  
                                      "AND t.user_id = 1 AND t.type_id = 2 AND strftime('%Y', t.date) = '2022' AND strftime('%m', t.date) = '10'" 
                                      "GROUP BY strftime('%d', t.date)", connect)
dfBarEXMonth = pd.DataFrame(monthlyBarExpense)

dayEx_total = dfBarEXMonth['Day'].values.tolist()
TypeEx_total = dfBarEXMonth['Type'].values.tolist()
AmountEx_total = dfBarEXMonth['Amount'].values.tolist()

type_total_combine = TypeIn_total + TypeEx_total
month_total_combine = dayIn_total + dayEx_total
amount_total_combine = AmountIn_total + AmountEx_total
df2 = pd.DataFrame(list(zip(month_total_combine, type_total_combine, amount_total_combine)), columns =['Day', 'Type', 'Amount'])

fig, ax = plt.subplots(figsize=(7, 5), dpi=100)
plot = sns.barplot(x='Day', y='Amount', data=df2, hue='Type', palette="Pastel1", ci=None)
for c in ax.containers:
    ax.bar_label(c, fmt='%.2f')
plot.set_ylabel("Amount (RM)", fontsize=15)
plot.set_xlabel("Day", fontsize=15)
plot.set_title("Income and Expense in a Month", fontsize=20)
plot.legend(bbox_to_anchor=(1,1), loc="upper right", borderaxespad=0)
plt.show()
```
